```python
# Задание 1

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1.0, 15.01, 0.01)
f = np.sin(x/5)*np.exp(x/10) + 5*np.exp(-x/2)

print('x:', x[0], 'f(1):', f[0])
print('x:', x[300], 'f(4):', f[300])
print('x:', x[700], 'f(8):', f[700])
print('x:', x[900], 'f(10):', f[900])
print('x:', x[1400], 'f(15):', f[1400])

a1 = np.array([[1, x[0]], 
               [1, x[1400]]])
b1 = np.array([f[0], 
               f[1400]])
print('Многочлен первой степени плохо приближает исходную функцию:', 'f =', np.linalg.solve(a1, b1)[1], 'x +', np.linalg.solve(a1, b1)[0])

a2 = np.array([[1, x[0], x[0]*x[0]],
               [1, x[700], x[700]*x[700]], 
               [1, x[1400], x[1400]*x[1400]]])
b2 = np.array([f[0], 
               f[700], 
               f[1400]])
print('Многочлен второй степени тоже плохо приближает исходную функцию:', 'f =', np.linalg.solve(a2, b2)[2], 'x*x +', np.linalg.solve(a2, b2)[1], 'x +', np.linalg.solve(a2, b2)[0])

a3 = np.array([[1, x[0], x[0]*x[0], x[0]*x[0]*x[0]],
               [1, x[300], x[300]*x[300], x[300]*x[300]*x[300]],
               [1, x[900], x[900]*x[900], x[900]*x[900]*x[900]], 
               [1, x[1400], x[1400]*x[1400], x[1400]*x[1400]*x[1400]]])
b3 = np.array([f[0], 
               f[300],
               f[900],
               f[1400]])
print('Многочлен третьей степени достойно приближает исходную функцию:', 'f =', np.linalg.solve(a3, b3)[3], 'x*x*x +', np.linalg.solve(a3, b3)[2], 'x*x +', np.linalg.solve(a3, b3)[1], 'x +', np.linalg.solve(a3, b3)[0])

plt.figure(figsize=(8,6))
plt.plot(x, f, 'r-', color='blue', label='функция')
plt.plot(x, np.linalg.solve(a1, b1)[1] * x + np.linalg.solve(a1, b1)[0], 'r-', color='red', label='многочлен первой степени')
plt.plot(x, np.linalg.solve(a2, b2)[2] * x * x + np.linalg.solve(a2, b2)[1] * x + np.linalg.solve(a2, b2)[0], 'r-', color='orange', label='многочлен второй степени')
plt.plot(x, np.linalg.solve(a3, b3)[3] * x * x * x + np.linalg.solve(a3, b3)[2] * x * x + np.linalg.solve(a3, b3)[1] * x + np.linalg.solve(a3, b3)[0], 'r-', color='green', label='многочлен третьей степени')
plt.legend()

print('Ответ: 4.36, -1.30, 0.19, -0.01.')
```
